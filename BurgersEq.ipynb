## Programa solución de la ecuación de Burgers

```python
%pylab inline
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
import numpy as np
```

### La primera posible solucion:

# $$  u_i^{j+1} = u_i^{j-1} - \beta \frac{(u^2)_{i+1}^j - (u^2)_{i-1}^j}{2} $$

donde $\beta$ equivale a $c\frac{\Delta t}{\Delta x}$. luego c es tomada como 1, y por ende $\beta = \frac{\Delta t}{\Delta x}$

```python
n_t = 110
n_x = 200

x = linspace(0,2,n_x)
t = linspace(0,0.15,n_t)
ui = 3*sin(pi*x)

dt = 0.0015
dx = x[1]-x[0]
print dt/dx


```

Se grafican en primer lugar los valores de la funcion sinusoidal inicial $3.0 sin(\pi x)$, con $ 0 < x < 2$ 

```python
plot (x,ui)
```

```python
#fixed boundary conditions
ui[0] = 0.0
ui[n_x/2] = 0.0

uf = zeros(n_x)
up = ui.copy()

un = ui.copy()

us = np.zeros(n_t*n_x).reshape(n_t, n_x)
us[0]=up

for j in range(1,n_t):
    for i in range(1,n_x-1):
        uf[i] = up[i] - dt/dx*((un[i+1])**2-(un[i-1])**2)/2
    us[j] = uf
    up = un.copy()
    un = uf.copy()
    
   
```

Luego se grafica un valor final de la grafica para la primera posible solución.

```python
plot (x,un)
```

Seguida por su respectiva grafica tridimensional donde se incluye el cambio en el tiempo

```python
x = linspace(0,2,n_x)
t = linspace(0,0.15,n_t)

fig = plt.figure(figsize(15,10))
ax = fig.gca(projection='3d')
x ,t = np.meshgrid(x,t)
surf = ax.plot_wireframe(x,t,us)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

plt.savefig("1.png")
plt.show()
```

## La segunda posible solución es:

### $$  u_i^{j+1} = u_i^{j} - \frac{\beta}{4} \left((u^2)^j_{i+1} - (u^2)^j_{i-1}\right) + \frac{\beta^2}{8} \left((u^j_{i+1} + u^j_i)((u^2)^j_{i+1} - (u^2)^j_{i-1}) - (u^j_i + u^j_{i-1})((u^2)^j_i - (u^2)^j_{i-1}) \right) $$

Bajo mismas condiciones de frontera, se busca tambien resolver la ecuacion de la segunda forma.

```python
n_t = 110
n_x = 200

x = linspace(0,2,n_x)
t = linspace(0,0.15,n_t)
ui = 3*sin(pi*x)

dt = 0.0015
dx = x[1]-x[0]
print dt/dx
```

```python
#fixed boundary conditions
ui[0] = 0.0
ui[n_x/2] = 0.0

uf = zeros(n_x)
un = ui.copy()
b = dt/dx

us2 = np.zeros(n_t*n_x).reshape(n_t, n_x)
us2[0]=un

for j in range(1,n_t):
    for i in range(1,n_x-1):
        uf[i] = un[i] - b/4*((un[i+1])**2-(un[i-1])**2)\
            + b**2/8 *((un[i+1]+un[i])*((un[i+1])**2-(un[i-1])**2)\
            - (un[i]+un[i-1])*((un[i+1])**2-(un[i-1])**2))
    us2[j] = uf
    un = uf.copy()
```

En el segundo caso se puede observar una amplitud mayor en el eje de $u$ que en la solucion del caso anterior, por lo que se podria tomar como mas preciso que el anterior.

```python
figsize(6,4)
plot (x, uf)
```

```python
x = linspace(0,2,n_x)
t = linspace(0,0.15,n_t)

fig = plt.figure(figsize(15,10))
ax = fig.gca(projection='3d')
x ,t = np.meshgrid(x,t)
surf = ax.plot_wireframe(x,t,us2)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

plt.savefig('2.png')
plt.show()
```

En la segunda solucion se puede ver que se aproxima mas rapido a los valores anomalos de ruido, luego, permite encontrar los picos mas facilmente a comparacion de la anterior solucion, debido a que esta hace uso de una expansion hasta de segundo orden de la funcion.
