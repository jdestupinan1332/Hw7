# $$  u_i^{j+1} = u_i^{j-1} - \beta \frac{(u^2)_{i+1}^j - (u^2)_{i-1}^j}{2} $$

### $$  u_i^{j+1} = u_i^{j} - \frac{\beta}{4} \left((u^2)^j_{i+1} - (u^2)^j_{i-1}\right) + \frac{\beta^2}{8} \left((u^j_{i+1} + u^j_i)((u^2)^j_{i+1} - (u^2)^j_{i-1}) - (u^j_i + u^j_{i-1})((u^2)^j_i - (u^2)^j_{i-1}) \right) $$

```python
n_t = 110
n_x = 200

x = linspace(0,2,n_x)
t = linspace(0,0.15,n_t)
ui = 3*sin(pi*x)

dt = 0.0015
dx = x[1]-x[0]
print dt/dx


```

```python
#fixed boundary conditions
ui[0] = 0.0
ui[n_x/2] = 0.0

uf = zeros(n_x)
un = ui.copy()
b = dt/dx

us2 = np.zeros(n_t*n_x).reshape(n_t, n_x)
us2[0]=un

for j in range(1,n_t):
    for i in range(1,n_x-1):
        uf[i] = un[i] - b/4*((un[i+1])**2-(un[i-1])**2)\
            + b**2/8 *((un[i+1]+un[i])*((un[i+1])**2-(un[i-1])**2)\
            - (un[i]+un[i-1])*((un[i+1])**2-(un[i-1])**2))
    us2[j] = uf
    un = uf.copy()
```

```python
plot (x, uf)
```

```python
x = linspace(0,2,n_x)
t = linspace(0,0.15,n_t)

fig = plt.figure()
ax = fig.gca(projection='3d')
x ,t = np.meshgrid(x,t)
surf = ax.plot_wireframe(x,t,us2)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

plt.show()
```
