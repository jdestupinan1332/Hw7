# $$   u_i^{j+1} = u_i^j - \beta \frac{(u^2)_i^j - (u^2)_{i-1}^j}{2} $$

```python
n_t = 10
n_x = 101
dt = 0.01
dx = x[1]-x[0]
print dt/dx

x = linspace(0,2,n_x)
ui = 3*sin(pi*x)
```

```python
plot (x,ui)
```

```python
#fixed boundary conditions
ui[0] = 0.0
ui[n_x/2] = 0.0

uf = zeros(n_x)

un = ui.copy()

for j in range(n_t):
    for i in range(1,n_x-1):
        uf[i] = un[i] - dt/dx*((un[i])**2-(un[i-1])**2)/2
        
    un = uf.copy()
```

```python
plot (x,un)
```

# $$  u_i^{j+1} = u_i^{j-1} - \beta \frac{(u^2)_{i+1}^j - (u^2)_{i-1}^j}{2} $$

```python
n_t = 10
n_x = 101
dt = 0.01
dx = x[1]-x[0]
print dt/dx

x = linspace(0,2,n_x)
ui = 3*sin(pi*x)
```

```python
plot (x,ui)
```

```python
#fixed boundary conditions
ui[0] = 0.0
ui[n_x/2] = 0.0

uf = zeros(n_x)
up = zeros(n_x)

un = ui.copy()

for j in range(n_t):
    for i in range(1,n_x-1):
        uf[i] = up[i] - dt/dx*((un[i+1])**2-(un[i-1])**2)/2
    uf[-1] = un[-1] - dt/dx*((un[-1])**2-(un[-2])**2)/2
    up = un.copy()
    un = uf.copy()
   
```

```python
plot (x,un)
```
